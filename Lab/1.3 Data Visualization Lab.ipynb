# Ranges for Quantitative Features

Using the visuals and description of the dataset I set the ranges for what "should" be a top hit +/- a given ammount based on size from the 50% mark. EX: Acousticness had a large 50-75% box plot so the max value was skewed more in that direction.

```python
criteria_mask = (
    (df['bpm'].between(115, 125)) &
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['danceability_%'].between(65, 75)) &
    (df['energy_%'].between(60, 70)) &
    (df['speechiness_%'].between(5, 9)) &
    (df['acousticness_%'].between(16, 24))
)

smash_hits = df[criteria_mask]

smash_hits_sorted = smash_hits.sort_values(by='streams', ascending=False)

print(f"Found {len(smash_hits)} songs that meet ALL criteria.")
print("\n--- Matching Songs ---")
# displaying specific columns
display_cols = ['track_name', 'artist(s)_name', 'streams', 'bpm', 'key', 'mode']
smash_hits_sorted[display_cols]
```

# Defining mathmatical ranges for Quantitative features

```python
hit_criteria = {}

# 1. Quantitative Features: Use the Interquartile Range (25th to 75th percentile)
quant_features = ['bpm', 'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']

print("--- Quantitative Criteria (Ranges) ---")
for feature in quant_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    hit_criteria[feature] = (Q1, Q3)
    print(f"{feature}: {Q1} to {Q3}")

# 2. Categorical Features: Use the Mode (Most common value)
cat_features = ['key', 'mode']

print("\n--- Categorical Criteria (Most Common) ---")
for feature in cat_features:
    most_common = df[feature].mode()[0]
    hit_criteria[feature] = most_common
    print(f"{feature}: {most_common}")
```

# Identifying Smash Hits

```python
mask = pd.Series([True] * len(df))

# Filter by Quantitative ranges
for feature in quant_features:
    min_val, max_val = hit_criteria[feature]
    mask = mask & (df[feature] >= min_val) & (df[feature] <= max_val)

# Filter by Categorical values
for feature in cat_features:
    target_val = hit_criteria[feature]
    mask = mask & (df[feature] == target_val)

# Create the result dataframe
potential_hits = df[mask]

print(f"Number of songs meeting ALL criteria: {len(potential_hits)}")
print("\nHere are the matches:")
display_cols = ['track_name', 'artist(s)_name', 'streams', 'bpm', 'key', 'mode']
potential_hits[display_cols]
```

# Compare known hit song, Miley Cirus - Flowers

```python
# Pick a known smash hit to check
song_name = "Flowers"
known_hit = df[df['track_name'].str.contains(song_name, case=False, na=False)]

if not known_hit.empty:
    hit_row = known_hit.iloc[0]
    print(f"Analysis for '{hit_row['track_name']}' by {hit_row['artist(s)_name']}:\n")
    
    # Check Quantitative features
    for feature in quant_features:
        val = hit_row[feature]
        low, high = hit_criteria[feature]
        status = "PASS" if low <= val <= high else "FAIL"
        print(f"{feature}: {val} (Target: {low}-{high}) -> {status}")
        
    # Check Categorical features
    for feature in cat_features:
        val = hit_row[feature]
        target = hit_criteria[feature]
        status = "PASS" if val == target else "FAIL"
        print(f"{feature}: {val} (Target: {target}) -> {status}")
else:
    print(f"Song '{song_name}' not found in the dataset.")
```

Going off these value ranges betweent the 25-75% marks, Flowers hits all features except for "Key".
